### **(2) 리스트 조회**

### **(3) 리스트 수정**

#### **1) 값 추가**
* 값을 추가하는 두가지 방법
    * ① .append()
    * ② 리스트 + [추가할 값] : 리스트1 + 리스트2 = 합쳐진 리스트

```python
a = [23, 3, 16, 45, 11]

a.append(99)
print(a)
```

```python
a.append([9, 10]) # element가 리스트여도 하나의 요소로 추가됨
print(a)
```

```python
a = [23, 3, 16, 45, 11]

a = a + [99]
print(a)
```

- 값을 추가하는 방법 another
  * ③ .extend()
  * ④ .insert()

```python
a = [23, 3, 16, 45, 11]
b = [-1, -5]

# append()와 달리, extend()는 리스트를 하나의 요소로 추가하는 것이 아닌, 각 요소를 개별적으로 추가함
a.extend(b)
print(a)
```

```python
a = [23, 3, 16, 45, 11]

a.insert(2, -999)
print(a)
```

요약하자면,

- append(element): 하나의 element를 리스트의 끝에 추가
- extend(iterable): iterable의 모든 요소를 리스트의 끝에 하나씩 추가
- insert(index, element): index 위치에 element를 삽입


#### **2) 값 수정**
* 인덱스로 해당 값을 조회하고 그 자리에 수정
    * a[2] = 34

```python
a = [23, 3, 16, 45, 11]

a[2] = 34
print(a)
```

#### **3) 값 삭제**
* 인덱스로 조회한 대상을 삭제 : del a[2]

```python
a = [23, 3, 16, 45, 11]

del a[2]
print(a)
```

* 특정 값을 찾아서 삭제 :
    * 리스트.remove(값)

```python
a = [23, 3, 16, 45, 11]

a.remove(45)
print(a)
```

## **2.딕셔너리(dictionary)**

* 하나의 딕셔너리 안에 다양한 형태의 여러 값들을 저장
* 여러 자료형들의 묶음

### **(1) 생성**

* 중괄호 {} 안에 , key : value 형태로 선언합니다.
    * Value : 다양한 형태의 변수
    * Key : Value 변수의 이름

```python
dict_a = { 'v1' : 32,
           'l1' : [1, 2, 3],
           'd1' : {'a' : 1, 'b' : 2} }
print(dict_a)
```

* 딕셔너리 정보 조회
    * .keys() : 딕셔너리의 key 만 조회
    * .values() : 딕셔너리의 값 만 조회
    * .items() : key와 값을 쌍(tuple)으로 조회

```python
dict_a.keys()
```

```python
dict_a.values()
```

```python
dict_a.items()
```

### **(2) 조회**

* 조회시 키로 값을 조회
    * 딕셔너리[키]

```python
dict_a = { 'v1': 32, 'l1': [1,2,3], 'd1': {'a':1, 'b':2}}

print( dict_a['v1'])
print('=' * 30)

print( dict_a['l1'])
print( dict_a['l1'][:2])
print('=' * 30)

print( dict_a['d1'])
print( dict_a['d1']['a'])
```

## **3.튜플(tuple)**

* 튜플은 좌표, 순서쌍을 담는 자료형입니다.

```python
# 소괄호로 선언
t = (4,7)
t
```

```python
# 조회는 인덱스로
print(t[0])
print(t[1])
```

```python
# 튜플의 값을 변경하려고 시도한다면 에러 발생
t[0] = 1
```

튜플은 불변(immuable) 자료형이기 때문에 값을 변경하려 하면 오류가 발생합니다. 튜플의 각 요소는 한 번 정의되면 변경할 수 없습니다. 튜플에 포함된 요소를 수정하거나, 새 요소를 추가하거나, 기존 요소를 삭제하는 등의 작업이 불가능합니다.

반면에, 리스트(list)는 가변(mutable) 자료형이므로 리스트의 요소를 추가, 수정, 삭제할 수 있습니다.

그렇다면 튜플의 요소를 변경하려면 어떻게 해야 할까요?

```python
# 리스트로 변환 후 수정
t = (4, 7)
temp_list = list(t)  # 튜플을 리스트로 변환
temp_list[0] = 1    # 리스트의 첫 번째 요소 변경
t = tuple(temp_list) # 리스트를 다시 튜플로 변환
print(t)  # 출력: (1, 7)
```

```python
# 새 튜플 생성
t = (4, 7)
t = (1,) + t[1:]  # 새로운 튜플 생성
print(t)  # 출력: (1, 7)
```
