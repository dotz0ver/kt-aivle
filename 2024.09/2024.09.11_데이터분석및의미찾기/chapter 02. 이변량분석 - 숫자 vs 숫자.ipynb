## **3.수치화 : 상관분석**

### (1) 상관계수, p-value

* 상관계수 𝑟
    * 공분산을 표준화 한 값
    * -1 ~ 1 사이의 값
    * -1, 1에 가까울 수록 강한 상관관계를 나타냄.

* 경험에 의한 대략의 기준(절대적인 기준이 절대 아닙니다.)
    * 강한 : 0.5 < |𝑟| ≤ 1
    * 중간 : 0.2 < |𝑟| ≤ 0.5
    * 약한 : 0.1 < |𝑟| ≤ 0.2
    * (거의)없음 : |𝑟| ≤ 0.1
    

```python
import scipy.stats as spst
```

```python
# 상관계수와 p-value
spst.pearsonr(air['Temp'], air['Ozone'])
```

결과는 튜플로 나오는데

1. 튜플의 첫 번째 값 : 상관계수를 뜻합니다.
2. 두번째 값 : p-value
    * 귀무가설 : 상관 관계가 없다.(상관계수가 0이다.)
    * 대립가설 : 상관 관계가 있다.(상관계수가 0이 아니다.)
3. 주의 사항 : 값에 NaN이 있으면 계산되지 않습니다. 반드시 .notnull()로 제외하고 수행해야 합니다.

### **(2) 데이터프레임 한꺼번에 상관계수 구하기**

```python
# 데이터프레임으로 부터 수치형 데이터에 대한 상관계수 구하기
air.corr()
```

위 결과로 부터,
1. 같은 변수끼리 구한 값 1은 의미 없다.
2. 상관계수의 절대값이
    * 1에 가까울 수록 강한 상관관계
    * 0에 가까울 수록 약한 상관관계
3. +는 양의 상관관계, -는 음의 상관관계

### **(3) (추가) 상관계수를 heatmap으로 시각화**

```python
plt.figure(figsize = (8, 8))
sns.heatmap(air.corr(),
            annot = True,            # 숫자(상관계수) 표기 여부
            fmt = '.3f',             # 숫자 포멧 : 소수점 3자리까지 표기
            cmap = 'RdYlBu_r',       # 칼라맵
            vmin = -1, vmax = 1)     # 값의 최소, 최대값
plt.show()
```

칼라를 변경하려면 아래 링크로 가서 color map 을 확인하고 조정하세요.

* 다음의 변수에 대해서 상관분석을 수행하시오.

```python
spst.pearsonr(air['Temp'], air['Ozone']) # 상관계수, pvalue
```

> * **dis** :	직업센터의 거리
> * **rad** :	방사형 고속도로까지의 거리
> * **tax** :	재산세율

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as spst
```

```python
# 보스톤 집값 데이터
boston = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/boston.csv')
boston.head()
```

* 1) crim(범죄율) --> mdev(집값) 에 대해 시각화와 수치화(상관분석)을 수행후 관계를 평가하시오.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x=boston['crim'], y=boston['medv'])
plt.title('Crime Rate vs House Price')
plt.xlabel('Crime Rate')
plt.ylabel('House Price (medv)')
```

```python
corr_crim_medv = boston['crim'].corr(boston['medv'])
```

```python
corr_crim_medv
```

범죄율(crim)과 집값(medv)의 관계:

- 시각화: 범죄율과 집값 사이에는 반비례 관계가 있는 것으로 보입니다. 범죄율이 높을수록 집값이 낮아지는 경향을 확인할 수 있습니다.
- 상관계수: 상관계수는 약 -0.39로, 이는 약한 음의 상관관계를 의미합니다. 즉, 범죄율이 높아질수록 집값이 낮아지는 경향이 있지만, 강한 관계는 아닙니다.

* 2) tax(제산세율) --> mdev(집값) 에 대해 시각화와 수치화(상관분석)을 수행후 관계를 평가하시오.

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=boston['tax'], y=boston['medv'])
plt.title('Property Tax Rate vs House Price')
plt.xlabel('Property Tax Rate')
plt.ylabel('House Price (medv)')
```

```python
corr_tax_medv = boston['tax'].corr(boston['medv'])
```

```python
corr_tax_medv
```

재산세율(tax)과 집값(medv)의 관계:
- 시각화: 재산세율과 집값 사이에도 음의 관계가 나타납니다. 재산세율이 높을수록 집값이 낮아지는 경향을 확인할 수 있습니다.
- 상관계수: 상관계수는 약 -0.47로, 이는 중간 정도의 음의 상관관계를 의미합니다. 즉, 재산세율이 높아질수록 집값이 더 낮아지는 경향이 있습니다.

* 3) lstat(하위계층비율) --> mdev(집값) 에 대해 시각화와 수치화(상관분석)을 수행후 관계를 평가하시오.

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x=boston['lstat'], y=boston['medv'])
plt.title('Lower Status Population Rate vs House Price')
plt.xlabel('Lower Status Population Rate (lstat)')
plt.ylabel('House Price (medv)')
```

```python
corr_lstat_medv = boston['lstat'].corr(boston['medv'])
```

```python
corr_lstat_medv
```

하위계층 비율(lstat)과 집값(medv)의 관계:
- 시각화: 하위계층 비율이 높아질수록 집값이 낮아지는 매우 강한 음의 관계가 시각적으로 확인됩니다.
- 상관계수: 상관계수는 약 -0.74로, 이는 매우 강한 음의 상관관계를 의미합니다. 즉, 하위계층 비율이 높아질수록 집값이 크게 낮아지는 경향이 있습니다.

* 4) 전체 변수들끼리의 상관계수를 구하고, 가장 강한 상관관계와 가장 약한 상관관계를 찾아 봅시다.

```python
corr_matrix = boston.corr()
```

```python
# 가장 강한 상관관계와 가장 약한 상관관계 찾기
strongest_corr = corr_matrix['medv'].drop('medv').idxmax(), corr_matrix['medv'].drop('medv').max()
weakest_corr = corr_matrix['medv'].drop('medv').idxmin(), corr_matrix['medv'].drop('medv').min()
```

```python
strongest_corr, weakest_corr
```

전체 변수들끼리의 상관관계:
- 가장 강한 상관관계: rm (주택당 방 수)와 medv의 상관계수는 0.695로, 방 수가 많을수록 집값이 높아지는 강한 양의 상관관계를 보여줍니다.
- 가장 약한 상관관계: lstat (하위계층 비율)과 medv의 상관계수는 -0.737로, 하위계층 비율이 높아질수록 집값이 낮아지는 가장 강한 음의 상관관계를 보여줍니다.
