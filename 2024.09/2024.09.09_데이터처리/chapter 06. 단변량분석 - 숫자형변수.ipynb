#### **4) 4분위수**

4분위수(四分位數, Quartiles)는 데이터를 작은 값에서 큰 값으로 정렬했을 때, 이를 네 개의 구간으로 나누는 기준이 되는 값들을 의미합니다. 각 구간은 데이터의 25%씩을 포함하며, 주로 데이터의 분포를 이해하고, 이상치(Outlier)를 식별할 때 유용하게 사용됩니다.

1사분위수 (Q1):
- 데이터의 하위 25%에 해당하는 값. 데이터를 정렬했을 때, 첫 번째 25% 지점에 위치하는 값입니다.
- 하위 사분위수라고도 하며, 25번째 백분위수와 같습니다.

2사분위수 (Q2):
- 데이터의 중앙값 (Median)으로, 데이터의 50% 지점에 위치하는 값입니다.
- 데이터의 중간값이며, 50번째 백분위수와 같습니다.

3사분위수 (Q3):
- 데이터의 상위 25%를 제외한 값으로, 상위 75% 지점에 위치하는 값입니다.
- 상위 사분위수라고도 하며, 75번째 백분위수와 같습니다.

4사분위수:
- 데이터의 최대값으로, 데이터의 상위 100% 지점에 위치하는 값입니다.
- 일반적으로 4사분위수는 최대값을 의미합니다.


4분위수 계산 방법:

데이터셋을 작은 값에서 큰 값으로 정렬한 후, 데이터를 네 개의 동일한 부분으로 나누는 값들을 찾습니다.

히스토그램과 어떤 차이가 있나요?

- 히스토그램: 연령대별 승객의 수를 나타낼 때, 특정 연령대에 승객이 얼마나 많은지 확인할 수 있습니다.
- 밀도 함수 그래프: 승객의 나이 분포가 특정 연령대에 집중되어 있는지, 나이 분포가 여러 군데에 걸쳐 분포하고 있는지를 부드러운 곡선으로 확인할 수 있습니다.

```python
sns.kdeplot(x = 'Age',
            data = titanic)

plt.show()
```

[문2] titanic Age에 대해서 히스토그램을 그려봅시다.
* 단, 이번에는 sns.histplot 을 이용합니다.
* 옵션으로 kde = True 를 지정해 봅시다.

```python
sns.histplot(x = 'Age',
             data = titanic,
             kde = True)
plt.show()
```

Sturges' Rule
- 히스토그램의 `bin` 개수를 계산하는 규칙입니다.
- 데이터의 개수 `n`을 기준으로 적절한 `bin` 개수를 결정합니다.

공식
$$
k = 1 + \log_2(n)
$$
- `k`: 히스토그램의 bin 개수
- `n`: 데이터의 개수

기본적으로 `sns.histplot()`에서 `bin`의 개수를 지정하지 않으면, 이 규칙을 사용하여 `bin`의 개수가 결정됩니다.

첫 번째 박스 플롯 (수평 방향):
- 중앙값 (Median): 박스 내부의 중앙선은 Fare 값의 중앙값(50번째 백분위수)을 나타냅니다. 이는 대부분의 승객들이 낮은 운임 구간에 속함을 보여줍니다.
- 박스: 박스의 상하단은 각각 1사분위수(Q1, 25%)와 3사분위수(Q3, 75%)를 나타내며, 이는 데이터의 중심 50%를 포함합니다.
- 수염 (Whiskers): 수염은 1.5 * IQR(Interquartile Range, 사분위 범위)을 기준으로 하여 상하로 뻗어 있으며, 이 범위 밖의 점들은 이상치로 간주됩니다.
- 이상치 (Outliers): 운임이 100 이상인 승객들은 대부분 이상치로 표시되며, 이는 극소수의 승객들이 매우 높은 운임을 지불했음을 나타냅니다.

```python
sns.boxplot(y = 'Fare',
            data = titanic)
plt.grid()
plt.show()
```

두 번째 박스 플롯 (수직 방향):
- 위와 동일한 정보를 수직 방향으로 표현한 그래프입니다. 대부분의 승객들이 0에서 100 사이의 운임을 지불했고, 특히 낮은 운임 구간에 많은 승객들이 집중되어 있음을 알 수 있습니다.
- 상단의 수많은 점들이 이상치를 나타내며, 500 이상을 지불한 승객은 매우 드물다는 것을 보여줍니다.

```python
sns.violinplot(x = 'Fare',
               data = titanic)
plt.grid()
plt.show()
```

바이올린 플롯은 박스 플롯과 밀도 함수를 결합하여, 데이터의 분포를 부드러운 곡선으로 나타냅니다.

- 밀도: Fare의 분포가 가장 높은 구간은 낮은 운임(약 0~50)에서 나타납니다. 이 구간에서 승객들이 몰려 있음을 보여줍니다.
- 대칭성: 왼쪽과 오른쪽이 대칭적으로 생겼으며, 대부분의 승객들이 낮은 운임 구간에 집중되어 있습니다.
- 이상치: 박스 플롯에서 나타난 이상치들이 여기에서도 보이며, 특히 500 이상의 운임 구간이 그래프 오른쪽 끝에 작은 꼬리처럼 나타나 있습니다.

## (2) 단변량 분석

* 단변량 분석 코드를 함수로 만들기(중요!)
    * 복잡하고 반복적인 코드를 함수로 만들어서 사용해 봅시다.

```python
def eda_1_n(data, var, bins = 30) :
    # 기초 통계량
    display(data[[var]].describe().T)

    # 시각화
    plt.figure(figsize = (12,8))
    plt.subplot(2,1,1)
    sns.histplot(data[var], bins = bins, kde = True)
    plt.grid()

    plt.subplot(2,1,2)
    sns.boxplot(x = data[var])
    plt.grid()
    plt.show()
```

* medv(집값)

```python
eda_1_n(boston, 'medv')
```

* crim(범죄율)

```python
eda_1_n(boston, 'crim')
```

* lstat(하위계층 비율)

```python
eda_1_n(boston, 'lstat')
```

* ptratio(교사1명당 학생수)

```python
eda_1_n(boston, 'ptratio')
```

* tax(재산세)

```python
eda_1_n(boston, 'tax')
```
